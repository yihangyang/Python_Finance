```python
data = pd.read_csv(filename,index_col = 0,parse_dates = True)
data.head()
```

```python
instruments = ['Apple Stock', 'Microsoft Stock',
               'Intel Stock', 'Amazon Stock', 'Goldman Sachs Stock',
               'SPDR S&P 500 ETF Trust', 'S&P 500 Index',
               'VIX Volatility Index', 'EUR/USD Exchange Rate',
               'Gold Price', 'VanEck Vectors Gold Miners ETF',
               'SPDR Gold Trust']
```

```python
for ric,name in zip(data.columns,instruments):
    print('{:8s} | {}'.format(ric,name))
```

```python
#Change Over Time
```

```python
data.diff().head()
```

```python
data.diff().mean()
```

```python
data.pct_change().round(3).head()
```

```python
data.pct_change().mean()
```

```python
data.pct_change().mean().plot(kind = 'bar',figsize = (10,6))
```

```python
rets = np.log(data/data.shift())
rets.head().round(3)
```

```python
rets.plot()
```

```python
rets.cumsum().apply(np.exp).plot()
```

```python
#Resampling
```

```python
data.resample('1w',label = 'right').last().head() #每次截止日期最后(每周日)
```

```python
data.resample('1m',label = 'right').last().head()
```

```python
rets.cumsum().apply(np.exp).resample('1m',label = 'right').last().plot()
```

```python
rets.cumsum().apply(np.exp).resample('1y',label = 'right').last().plot()
```
