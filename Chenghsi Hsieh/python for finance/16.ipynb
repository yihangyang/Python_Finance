```python
import numpy as np
def func(x):
    return 0.5 * np.exp(x) + 1

a,b = 0.5,1.5 #The interal limits
x = np.linspace(0,2) #The x values to plot the function...预设50点，切出49份
y = func(x)
Ix = np.linspace(a,b) #The x values within the integral limits从0.5到1.5等分49份
Iy = func(Ix)
```

```python
Ix
```

```python
Iy
```

```python
list(zip(Ix,Iy))
```

```python
verts = [(a,0)] + list(zip(Ix,Iy)) + [(b,0)]
verts
```

```python
len(verts)
```

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots(figsize=(10,6))

plt.plot(x,y,'b',linewidth = 2)
plt.ylim(bottom = 0)
```

```python
from matplotlib.patches import Polygon # 画多边形

fig,ax = plt.subplots(figsize = (10,6))

poly = Polygon([(0.5,0),(0.6,0),(0.6,func(0.6)-1),(0.5,func(0.5)-1)])

ax.add_patch(poly)
```

```python
fig,ax = plt.subplots(figsize = (10,6))

poly = Polygon([(0.5,0),(0.6,0),(0.5,func(0.5)-1),(0.6,func(0.6)-1)])

ax.add_patch(poly)
plt.grid()
```

```python
#big master
from matplotlib.patches import Polygon

fig,ax = plt.subplots(figsize=(10,6))

plt.plot(x,y,'b',linewidth = 2)
plt.ylim(bottom = 0)

poly = Polygon(verts,facecolor = '0.7',edgecolor = '0.5')
ax.add_patch(poly)

plt.text(0.5 * (a + b),1,r'$\int_a^b f(x)\mathrm{d}x$',
         horizontalalignment='center',fontsize = 20 )

plt.figtext(0.9,0.075,'$x$')
plt.figtext(0.075,0.9,'$f(x)$')

ax.set_xticks((a,b))
ax.set_xticklabels(('$a$','$b$'))

ax.set_yticks((func(a),func(b)))
ax.set_yticklabels(('$f(a)$','$f(b)$'))
```

```python
# static 3D Plotting
```

```python
strike = np.linspace(50,150,24)
strike
```

```python
ttm = np.linspace(0.5,2.5,24)
ttm
```

```python
strike,ttm = np.meshgrid(strike,ttm)
```

```python
strike
```

```python
strike.shape
```

```python
ttm
```

```python
ttm.shape
```

```python
strike[:2].round(1)
```

```python
iv = (strike -100) ** 2 / (100 * strike) / ttm
```

```python
iv.shape
```

```python
iv[:5,:3]
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,6))
ax = fig.gca(projection = '3d')

surf = ax.plot_surface(strike,ttm,iv,rstride = 2,cstride = 2,
                       cmap = plt.cm.coolwarm,linewidth = 0.5,
                      antialiased = True)#平滑化

ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')

fig.colorbar(surf,shrink = 0.5,aspect = 5)
```

```python
fig = plt.figure(figsize = (10,6))
ax = fig.add_subplot(111,projection = '3d')

ax.view_init(30,60)
ax.scatter(strike,ttm,iv,zdir = 'z',s = 25,c = 'b',marker = '^')

ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')
```
